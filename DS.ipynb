```python
# Calculate BMI of the players
df['Height(m)']=df['Height(ft)']*0.3048
```

```python
# Creating BMI Colums
df['BMI']=df['Weight(kgs)']/(df['Height(m)']**2)
```

```python
'''
Create a new columns with the new "BMI Evaluation"
Assign the BMI status based on the following condition
BMI <= 18.4 - "Underweight"
BMI <= 24.9 - "Healthy"
BMI <= 29.9 - "Overweight"
BMI <= 34.9 - "Severly Over weight"
BMI <= 39.9  - "Obese"
BMI > 39.9 - "Severly Obese"
'''
# Create an empty list to store categories
BMI_Category=[]
# Loop through BMI Values
for bmi_value in df['BMI']:
  if bmi_value <=18.4:
    BMI_Category.append("Underweight")
  elif bmi_value <=24.9:
    BMI_Category.append("Healthy")
  elif bmi_value <=29.9:
    BMI_Category.append("Overweight")
  elif bmi_value <=34.9:
    BMI_Category.append("Severly Over weight")
  elif bmi_value <=39.9:
    BMI_Category.append("Obese")
  else:
    BMI_Category.append("Severly Obese")
df['BMI Evaluation']=BMI_Category
df
```
